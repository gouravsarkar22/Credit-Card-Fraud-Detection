## Feature engineering: Transaction hour
"Time" is seconds since first transaction in dataset

```python
df['Transaction_Hour'] = (df['Time'] // 3600) % 24
```

## Amount bins

```python
df['Amount_bin'] = pd.cut(df['Amount'], bins=[-1,10,50,200,1000, df['Amount'].max()], labels=['0-10','10-50','50-200','200-1000','1000+'])
```

## EDA plots

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Class', data=df)
plt.title('Class distribution (0=legit, 1=fraud)')
```

## Amount distribution

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale('symlog')
plt.title('Amount by Class (log scale)')
```

## Fraud by hour

```python
fraud_by_hour = df.groupby('Transaction_Hour')['Class'].mean()
plt.figure(figsize=(10,4))
fraud_by_hour.plot(kind='bar')
plt.title('Fraud rate by transaction hour')
```

## Class ratio

```python
print('Fraud ratio:', df['Class'].mean())
```

## Interpretation notes for EDA:

The dataset is extremely imbalanced (fraud << legit). Typical fraud ratio ~0.17%.

Visualizing Amount vs Class often shows frauds concentrated in higher amounts, but distribution overlaps.

Adding Transaction_Hour can reveal time-of-day patterns (e.g., late-night spikes).
